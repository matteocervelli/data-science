# Same data, different outcome

```r
install.packages('Tmisc')
```

```r
library(Tmisc)
library(dplyr)
library(ggplot2)
```

```r
data(quartet)
View(quartet)
```

We notice there are 4 sets of x and y variables.

```r
quartet %>%
  group_by(set) %>%
  summarize(mean(x), sd(x), mean(y),sd(y),cor(x,y))
```

As we can see, the mean, standard deviation, and correlation practically the same for each set.

But looking at the summarized data, it is not good to make any conclusion about the relationship between x and y.

Let's look at the data in a plot.

```r
ggplot(quartet, aes(x,y)) +
  geom_point() +
  geom_smooth(method = 'lm', se = FALSE) +
  facet_wrap(~set)
```
